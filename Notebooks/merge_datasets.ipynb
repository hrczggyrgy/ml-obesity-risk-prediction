```python
# First, let's load the required dataset using pandas, concatenate them into one and check the first few rows
import pandas as pd

df1 = pd.read_csv('/work/ObesityDataSet.csv')
df2 = pd.read_csv('/work/train.csv')

# Concatenate the two datasets
merged_df = pd.concat([df1, df2], ignore_index=True)

# Write the merged DataFrame into a CSV file
merged_df.to_csv('/work/merged_dataset.csv', index=False)

merged_df.head()
```
